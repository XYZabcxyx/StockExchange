# Data visualitation

```python
import matplotlib.pyplot as plt
```

```python
#displays the columns of the dataset
data.columns
```

```python
#train the data
df = pd.DataFrame(data)
```

```python
#Generate the pie chart
df = pd.DataFrame(data.head(20))
df['Low'].value_counts().plot(kind='pie',autopct='%1.1f%%')
plt.title("Low stock rate")
plt.show()
```

```python
df = pd.DataFrame(data.head(20))
df['High'].value_counts().plot(kind='pie',autopct='%1.1f%%')
plt.title("High stock rate")
plt.show()
```

```python
#generate the Bar chat
df = pd.DataFrame(data.head(20))
df['High'].value_counts().plot(kind='bar',color='skyblue')
plt.xlabel('High')
plt.ylabel('count')
plt.title("High")
plt.show()
```

```python
#bar graph
df = pd.DataFrame(data.head(20))
df['Low'].value_counts().plot(kind='bar',color='Green')
plt.xlabel('Low')
plt.ylabel('Count')
plt.title("Low")
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better visibility
plt.tight_layout()
plt.show()
```

```python
#line graph
df = pd.DataFrame(data.head(20))
plt.figure(figsize=(8, 6))
df['Adj Close'].plot(kind='line', color='red', marker='o')
plt.xlabel('Adj Close')
plt.xlabel('Close')
plt.title("Adj close vs close")
plt.show()
```

```python
#generate the scatter plot
df = pd.DataFrame(data.head(30))
plt.scatter(df.index, df['High'],df['Low'], alpha=0.7, edgecolors='black')
plt.title('High stock price  vs low stock price')
plt.xlabel('High')
plt.ylabel('Low')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#histogram
df = pd.DataFrame(data.head(20))
df['Volume'].plot(kind='hist',color='blue',bins=10)
plt.xlabel('Volume')
plt.ylabel('Frequency')
plt.title("Volume Histogram")
#plt.grid(True)
plt.tight_layout()
plt.show()
```
